## Prewitt filters

**<a id="#">Task 4.1.1.</a>** Load the image `data/astronaut.png` and show it:

**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical 3&times;3 Prewitt filters:

```python
def prewitt_h(img_in):
    result = numpy.zeros(img_in.shape)
    size = 3
    n = (size - 1) // 2
    for p in numpy.ndindex(img_in.shape):
        neighborhood_sum = 0
        for q in numpy.ndindex(size, size):
            x = q[1]+p[1]-n
            y = q[0]+p[0]-n
            
            if x < img_in.shape[1] and x >=0 and y < img_in.shape[0] and y >=0 :
                if q[1] == 0:
                    neighborhood_sum += img_in[y, x] * -1
                elif q[1] == 1:
                    neighborhood_sum += img_in[y,x] * 0
                elif q[1] == 2:
                    neighborhood_sum += img_in[y,x] * 1
            else:
                neighborhood_sum = 0
                break

        result[p] = neighborhood_sum / 6

    return result
    
def prewitt_v(img_in):
    result = numpy.zeros(img_in.shape)
    size = 3
    n = (size - 1) // 2
    for p in numpy.ndindex(img_in.shape):
        neighborhood_sum = 0
        for q in numpy.ndindex(size, size):
            x = q[1]+p[1]-n
            y = q[0]+p[0]-n
            
            if x < img_in.shape[1] and x >=0 and y < img_in.shape[0] and y >=0 :
                if q[0] == 0:
                    neighborhood_sum += img_in[y, x] * -1
                elif q[0] == 1:
                    neighborhood_sum += img_in[y,x] * 0
                elif q[0] == 2:
                    neighborhood_sum += img_in[y,x] * 1
            else:
                neighborhood_sum = 0
                break

        result[p] = neighborhood_sum / 6

    return result
```

**<a id="#">Task 4.1.3.</a>** Test your above solution using the functions `prewitt_h` and `prewitt_v` for the previously loaded image and show the result:

```python
img_h = prewitt_h(img)
plt.figure(figsize=(15,8))
plt.imshow(img_h)
plt.colorbar()

img_v = prewitt_v(img)
plt.figure(figsize=(15,8))
plt.imshow(img_v)
plt.colorbar()
```

```python
import skimage.io
img_h_o = skimage.io.imread('data/astronaut_prewitt_h.tiff')
img_v_o = skimage.io.imread('data/astronaut_prewitt_v.tiff')
```
